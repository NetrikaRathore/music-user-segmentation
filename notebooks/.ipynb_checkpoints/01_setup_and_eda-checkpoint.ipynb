```python
# Cell 1 — Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 50)
pd.set_option('display.float_format', '{:.2f}'.format)

print("All libraries loaded successfully")
```

```python
# Cell 2 — Load raw files
music    = pd.read_csv("../data/raw/Music Info.csv")
listening = pd.read_csv("../data/raw/User Listening History.csv")

print("=== Music Info ===")
print(f"Shape   : {music.shape}")
print(f"Columns : {list(music.columns)}\n")

print("=== Listening History ===")
print(f"Shape   : {listening.shape}")
print(f"Columns : {list(listening.columns)}")
```

```python
# Cell 3 — Basic stats
print("=== Listening History ===")
print(f"Total play events : {listening.shape[0]:,}")
print(f"Unique users      : {listening.user_id.nunique():,}")
print(f"Unique tracks     : {listening.track_id.nunique():,}")
print(f"Playcount range   : {listening.playcount.min()} → {listening.playcount.max()}")

print("\n=== Music Info ===")
print(f"Total tracks : {music.shape[0]:,}")
print(f"Genres       : {music.genre.nunique()} unique")
print(f"Year range   : {music.year.min()} → {music.year.max()}")
print(f"Missing genre: {music.genre.isnull().sum()} rows")
```

```python
# Cell 5 — Playcount distribution per user
user_totals = listening.groupby('user_id')['playcount'].sum()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(user_totals, bins=50, color='steelblue', edgecolor='white')
axes[0].set_title('Total plays per user')
axes[0].set_xlabel('Total playcount')
axes[0].set_ylabel('Number of users')

axes[1].hist(np.log1p(user_totals), bins=50, color='coral', edgecolor='white')
axes[1].set_title('Total plays per user (log scale)')
axes[1].set_xlabel('Log(Total playcount)')
axes[1].set_ylabel('Number of users')

plt.tight_layout()
plt.savefig('../data/processed/plot_playcount_distribution.png', dpi=120)
plt.show()
print(user_totals.describe())
```

```python
# Cell 6 — Top 15 artists by total plays
top_artists = (
    listening.merge(music[['track_id', 'artist']], on='track_id', how='left')
    .groupby('artist')['playcount'].sum()
    .sort_values(ascending=False)
    .head(15)
)

plt.figure(figsize=(10, 5))
top_artists.plot(kind='bar', color='steelblue', edgecolor='white')
plt.title('Top 15 artists by total plays')
plt.ylabel('Total playcount')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('../data/processed/plot_top_artists.png', dpi=120)
plt.show()
```

```python
# Cell 7 — Genre distribution
genre_counts = music['genre'].value_counts().head(12)

plt.figure(figsize=(8, 4))
genre_counts.plot(kind='barh', color='mediumseagreen', edgecolor='white')
plt.title('Track count by genre (top 12)')
plt.xlabel('Number of tracks')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('../data/processed/plot_genre_distribution.png', dpi=120)
plt.show()
```

```python
# Cell 8 — Audio feature overview
audio_features = ['danceability', 'energy', 'valence',
                  'acousticness', 'speechiness', 'instrumentalness']

fig, axes = plt.subplots(2, 3, figsize=(13, 7))
axes = axes.flatten()

for i, feat in enumerate(audio_features):
    axes[i].hist(music[feat].dropna(), bins=40,
                 color='mediumpurple', edgecolor='white')
    axes[i].set_title(feat)
    axes[i].set_xlabel('Value (0–1)')

plt.suptitle('Audio feature distributions across all tracks', fontsize=13, y=1.01)
plt.tight_layout()
plt.savefig('../data/processed/plot_audio_features.png', dpi=120)
plt.show()
```

```python
# Cell 9 — First business insight: engagement split
p25 = user_totals.quantile(0.25)
low  = (user_totals < p25).sum()
high = (user_totals >= p25).sum()

print(f"25th percentile threshold : {p25:.0f} plays")
print(f"Low engagement users      : {low:,}  ({low/len(user_totals)*100:.1f}%)")
print(f"High engagement users     : {high:,} ({high/len(user_totals)*100:.1f}%)")
```
