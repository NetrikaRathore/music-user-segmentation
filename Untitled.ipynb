```python
import os

folders = [
    "data/raw",
    "data/processed",
    "notebooks",
    "src",
    "dashboard"
]

for folder in folders:
    os.makedirs(folder, exist_ok=True)

print("Folders created:")
for folder in folders:
    print(f"  ✓ {folder}")
```

```python
files = [
    "src/features.py",
    "src/model.py",
    "README.md",
    ".gitignore"
]

for file in files:
    if not os.path.exists(file):
        open(file, 'w').close()

# Write gitignore content
with open(".gitignore", "w") as f:
    f.write("data/\nvenv/\n__pycache__/\n*.pyc\n.DS_Store\n*.zip\n")

print("Files created ✓")
```

```python
for root, dirs, files in os.walk("."):
    # skip hidden and venv folders
    dirs[:] = [d for d in dirs if d not in ['venv', '.git', '__pycache__']]
    level = root.replace(".", "").count(os.sep)
    indent = "  " * level
    print(f"{indent}{os.path.basename(root)}/")
    for file in files:
        print(f"{indent}  {file}")
```
